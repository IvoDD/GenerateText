```python
import tensorflow as tf
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open("text_data.pickle", "rb") as file:
    text_data = pickle.load(file)
```

```python
all_text = ''.join(text_data)
```

```python
id2char = sorted(set(all_text))
len(id2char)
```

```python
id2char
```

```python
char2id = {c:i for i, c in enumerate(id2char)}
```

```python
occurrences = [0]*len(id2char)
for i, article in enumerate(text_data):
    if i%1000==0: print(i)
    for i in range(len(id2char)):
        if id2char[i] in article:
            occurrences[i]+=1
        
occurrences
```

```python
len(list(filter(lambda x: x>=50, occurrences)))
```

```python
not_want = set()
for i in range(len(id2char)):
    if occurrences[i]<50:
        not_want.add(id2char[i])
        
len(not_want)
```

```python
new_text_data = []
for i, article in enumerate(text_data):
    if i%1000==0: print(i)
    if len(not_want.intersection(set(article)))==0:
        new_text_data.append(article)
        
len(new_text_data)
```

```python
with open("text_data.pickle", "wb") as file:
    pickle.dump(new_text_data, file)
```
