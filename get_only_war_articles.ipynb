```python
import xml.etree.ElementTree as ET
root = ET.parse('bgwiki-20180920-corpus.xml').getroot()
```

```python
for article in root:
    if 'Война' in article.attrib['name'] or 'война' in article.attrib['name']:
        print(article.attrib['name'])
```

```python
def innertext(tag):
    return (tag.text or '') + ''.join(innertext(e) for e in tag) + (tag.tail or '')

def text_from_content(content):
    return ''.join([innertext(p)+'\n' for p in content])
```

```python
war_articles = []
for article in root:
    if 'Война' in article.attrib['name'] or 'война' in article.attrib['name']:
        war_articles.append(article)
        
len(war_articles)
```

```python
war_texts = [text_from_content(art.find('content')) for art in war_articles]
war_texts[:5]
```

```python
war_all = ''.join(war_texts)
len(war_all)
```

```python
len(set(war_all))
```

```python
bul_symbols = "АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЬЪЮЯабвгдежзийклмнопрстуфхцчшщьъюя"
len(bul_symbols)
```

```python
def split_text(txt):
    ans = []
    curr = ''
    for c in txt:
        if c in bul_symbols:
            curr += c
        else:
            if curr!='': ans.append(curr)
            ans.append(c)
            curr = ''
    if curr!='': ans.append(curr)
    return ans

split_text(war_texts[0])
```

```python
len(split_text(war_all))
```
