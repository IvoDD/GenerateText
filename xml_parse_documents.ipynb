```python
import xml.etree.ElementTree as ET
tree = ET.parse('bgwiki-20180920-corpus.xml')
```

```python
root = tree.getroot()
root.tag
```

```python
print(f"Num of articles: {len(root)}")
```

```python
content = root[1000].find('content')
content
```

```python
ET.tostring(content, encoding='unicode')
```

```python
def innertext(tag):
    return (tag.text or '') + ''.join(innertext(e) for e in tag) + (tag.tail or '')

for p in content:
    print(innertext(p))
```

```python
def text_from_content(content):
    return ''.join([innertext(p)+'\n' for p in content])

print(text_from_content(content))
```

```python
text_data = [text_from_content(child.find('content')) for child in root[:10]]

text_data
```

```python
text_data = [text_from_content(child.find('content')) for child in root]
```

```python
import pickle

with open("text_data.pickle", "wb") as file:
    pickle.dump(text_data, file)
```
